```python
def fillNullWithDefault(dataFrame):
    dataFrame.Trim[dataFrame.Trim.isnull()] = 'Bas'
    dataFrame.Transmission[dataFrame.Transmission.isnull()] = 'AUTO'
    dataFrame.SubModel[dataFrame.SubModel.isnull()] = 'UNKNOWN'
    
    return dataFrame
```

```python
def fillNullWithDefault_2(dataFrame):
    dataFrame.Trim[dataFrame.Trim.isnull()] = 'Bas'
    dataFrame.Transmission[dataFrame.Transmission.isnull()] = 'AUTO'
    
    return dataFrame
```

```python
def fillNullWithDefault_3(dataFrame):
    dataFrame.Trim[dataFrame.Trim.isnull()] = 'Bas'
    dataFrame.Transmission[dataFrame.Transmission.isnull()] = 'AUTO'
    dataFrame.SubModel[dataFrame.SubModel.isnull()] = 'UNKNOWN'
    dataFrame.PRIMEUNIT[dataFrame.PRIMEUNIT.isnull()] = 'NO'
    dataFrame.AUCGUART[dataFrame.AUCGUART.isnull()] = 'GREEN'
    
    return dataFrame
```

```python
def fillNullWithMedian(dataFrame):
    dataFrame.MMRAcquisitionAuctionAveragePrice[dataFrame.MMRAcquisitionAuctionAveragePrice.isnull()] = dataFrame.median().MMRAcquisitionAuctionAveragePrice
    dataFrame.MMRAcquisitionAuctionCleanPrice[dataFrame.MMRAcquisitionAuctionCleanPrice.isnull()] = dataFrame.median().MMRAcquisitionAuctionCleanPrice
    dataFrame.MMRAcquisitionRetailAveragePrice[dataFrame.MMRAcquisitionRetailAveragePrice.isnull()] = dataFrame.median().MMRAcquisitionRetailAveragePrice
    dataFrame.MMRAcquisitonRetailCleanPrice[dataFrame.MMRAcquisitonRetailCleanPrice.isnull()] = dataFrame.median().MMRAcquisitonRetailCleanPrice

    dataFrame.MMRCurrentAuctionAveragePrice[dataFrame.MMRCurrentAuctionAveragePrice.isnull()] = dataFrame.median().MMRCurrentAuctionAveragePrice
    dataFrame.MMRCurrentAuctionCleanPrice[dataFrame.MMRCurrentAuctionCleanPrice.isnull()] = dataFrame.median().MMRCurrentAuctionCleanPrice
    dataFrame.MMRCurrentRetailAveragePrice[dataFrame.MMRCurrentRetailAveragePrice.isnull()] = dataFrame.median().MMRCurrentRetailAveragePrice
    dataFrame.MMRCurrentRetailCleanPrice[dataFrame.MMRCurrentRetailCleanPrice.isnull()] = dataFrame.median().MMRCurrentRetailCleanPrice
    
    return dataFrame
```

```python
def fillNullWithMax(dataFrame):
    MaxWheelType = dataFrame.groupby('WheelTypeID').count()['RefId']
    dataFrame.WheelTypeID[dataFrame.WheelTypeID.isnull()] = MaxWheelType[MaxWheelType == MaxWheelType.max()].index[0]

    MaxNationality = dataFrame.groupby('Nationality').count()['RefId']
    dataFrame.Nationality[dataFrame.Nationality.isnull()] = MaxNationality[MaxNationality == MaxNationality.max()].index[0]

    MaxSize = dataFrame.groupby('Size').count()['RefId']
    dataFrame.Size[dataFrame.Size.isnull()] = MaxSize[MaxSize == MaxSize.max()].index[0]

    MaxTopThree = dataFrame.groupby('TopThreeAmericanName').count()['RefId']
    dataFrame.TopThreeAmericanName[dataFrame.TopThreeAmericanName.isnull()] = MaxTopThree[MaxTopThree == MaxTopThree.max()].index[0]
    
    return dataFrame
```

```python
def encodeCategories(dataFrame):
    from sklearn.preprocessing import LabelEncoder
    label = LabelEncoder()

    label.fit(dataFrame.Transmission.drop_duplicates())
    dataFrame.Transmission = label.transform(dataFrame.Transmission) 

    label.fit(dataFrame.Auction.drop_duplicates())
    dataFrame.Auction = label.transform(dataFrame.Auction) 

    label.fit(dataFrame.Make.drop_duplicates())
    dataFrame.Make = label.transform(dataFrame.Make) 

    label.fit(dataFrame.Model.drop_duplicates())
    dataFrame.Model = label.transform(dataFrame.Model) 

    label.fit(dataFrame.Trim.drop_duplicates())
    dataFrame.Trim = label.transform(dataFrame.Trim) 

    label.fit(dataFrame.SubModel.drop_duplicates())
    dataFrame.SubModel = label.transform(dataFrame.SubModel)

    label.fit(dataFrame.Nationality.drop_duplicates())
    dataFrame.Nationality = label.transform(dataFrame.Nationality)

    label.fit(dataFrame.Size.drop_duplicates())
    dataFrame.Size = label.transform(dataFrame.Size)

    label.fit(dataFrame.TopThreeAmericanName.drop_duplicates())
    dataFrame.TopThreeAmericanName = label.transform(dataFrame.TopThreeAmericanName)
    
    return dataFrame
```

```python
def encodeCategories_2(dataFrame):
    from sklearn.preprocessing import LabelEncoder
    label = LabelEncoder()

    label.fit(dataFrame.Transmission.drop_duplicates())
    dataFrame.Transmission = label.transform(dataFrame.Transmission) 

    label.fit(dataFrame.Auction.drop_duplicates())
    dataFrame.Auction = label.transform(dataFrame.Auction) 

    label.fit(dataFrame.Make.drop_duplicates())
    dataFrame.Make = label.transform(dataFrame.Make) 

    label.fit(dataFrame.Model.drop_duplicates())
    dataFrame.Model = label.transform(dataFrame.Model) 

    label.fit(dataFrame.Trim.drop_duplicates())
    dataFrame.Trim = label.transform(dataFrame.Trim)

    label.fit(dataFrame.Nationality.drop_duplicates())
    dataFrame.Nationality = label.transform(dataFrame.Nationality)

    label.fit(dataFrame.Size.drop_duplicates())
    dataFrame.Size = label.transform(dataFrame.Size)

    label.fit(dataFrame.TopThreeAmericanName.drop_duplicates())
    dataFrame.TopThreeAmericanName = label.transform(dataFrame.TopThreeAmericanName)
    
    return dataFrame
```

```python
def encodeCategories_3(dataFrame):
    from sklearn.preprocessing import LabelEncoder
    label = LabelEncoder()

    label.fit(dataFrame.Transmission.drop_duplicates())
    dataFrame.Transmission = label.transform(dataFrame.Transmission) 

    label.fit(dataFrame.Auction.drop_duplicates())
    dataFrame.Auction = label.transform(dataFrame.Auction) 

    label.fit(dataFrame.Make.drop_duplicates())
    dataFrame.Make = label.transform(dataFrame.Make) 

    label.fit(dataFrame.Model.drop_duplicates())
    dataFrame.Model = label.transform(dataFrame.Model) 

    label.fit(dataFrame.Trim.drop_duplicates())
    dataFrame.Trim = label.transform(dataFrame.Trim) 

    label.fit(dataFrame.SubModel.drop_duplicates())
    dataFrame.SubModel = label.transform(dataFrame.SubModel)

    label.fit(dataFrame.Nationality.drop_duplicates())
    dataFrame.Nationality = label.transform(dataFrame.Nationality)

    label.fit(dataFrame.Size.drop_duplicates())
    dataFrame.Size = label.transform(dataFrame.Size)

    label.fit(dataFrame.TopThreeAmericanName.drop_duplicates())
    dataFrame.TopThreeAmericanName = label.transform(dataFrame.TopThreeAmericanName)
    
    label.fit(dataFrame.PRIMEUNIT.drop_duplicates())
    dataFrame.PRIMEUNIT = label.transform(dataFrame.PRIMEUNIT)
    
    label.fit(dataFrame.AUCGUART.drop_duplicates())
    dataFrame.AUCGUART = label.transform(dataFrame.AUCGUART)
    
    return dataFrame
```

```python
def drop(dataFrame):
    dataFrame = dataFrame.drop(['Color', 'PurchDate', 'WheelType', 'PRIMEUNIT', 'AUCGUART', 'VNST'], axis=1)
    
    return dataFrame
```

```python
def drop_2(dataFrame):
    dataFrame = dataFrame.drop(['Color', 'PurchDate', 'WheelType', 'PRIMEUNIT', 'AUCGUART', 'VNST', 'SubModel'], axis=1)
    
    return dataFrame
```

```python
def drop_3(dataFrame):
    dataFrame = dataFrame.drop(['Color', 'PurchDate', 'WheelType', 'VNST'], axis=1)
    
    return dataFrame
```

```python
def prepareDataFrame(dataFrame):
    dataFrame = fillNullWithDefault(dataFrame)
    
    dataFrame = fillNullWithMedian(dataFrame)
    
    dataFrame = fillNullWithMax(dataFrame)
    
    dataFrame = encodeCategories(dataFrame)
    
    dataFrame = drop(dataFrame)
    
    return dataFrame
```

```python
def prepareDataFrame_2(dataFrame):
    dataFrame = fillNullWithDefault_2(dataFrame)
    
    dataFrame = fillNullWithMedian(dataFrame)
    
    dataFrame = fillNullWithMax(dataFrame)
    
    dataFrame = encodeCategories_2(dataFrame)
    
    dataFrame = drop_2(dataFrame)
    
    return dataFrame
```

```python
def prepareDataFrame_3(dataFrame):
    dataFrame = fillNullWithDefault_3(dataFrame)
    
    dataFrame = fillNullWithMedian(dataFrame)
    
    dataFrame = fillNullWithMax(dataFrame)
    
    dataFrame = encodeCategories_3(dataFrame)
    
    dataFrame = drop_3(dataFrame)
    
    return dataFrame
```

```python
train = prepareDataFrame_3(train)
```

```python
test = prepareDataFrame_3(test)
```

```python
from sklearn import cross_validation
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve, auc
import pylab as pl
```

```python
train_target = train.IsBadBuy
train_features = train.drop(['IsBadBuy', 'RefId'], axis=1)
test = test.drop(['RefId'], axis=1)
kfold = 5
itog_val = {}
```

```python
model_rfc = RandomForestClassifier(n_estimators = 80)
```

```python
ROCtrainTRN, ROCtestTRN, ROCtrainTRG, ROCtestTRG = cross_validation.train_test_split(train_features, train_target, test_size=0.25)
```

```python
pl.clf()
plt.figure(figsize=(8,6))
probas = model_rfc.fit(ROCtrainTRN, ROCtrainTRG).predict_proba(ROCtestTRN)
fpr, tpr, thresholds = roc_curve(ROCtestTRG, probas[:, 1])
roc_auc  = auc(fpr, tpr)
pl.plot(fpr, tpr, label='%s ROC (area = %0.2f)' % ('RandonForest',roc_auc))
pl.plot([0, 1], [0, 1], 'k--')
pl.xlim([0.0, 1.0])
pl.ylim([0.0, 1.0])
pl.xlabel('False Positive Rate')
pl.ylabel('True Positive Rate')
pl.legend(loc=0, fontsize='small')
pl.show()
```

```python
model_rfc.fit(train_features, train_target)
result.insert(1,'IsBadBuy', model_rfc.predict(test))
result.to_csv('ans.csv', index=False)
```
